```python
# Define the base working directory
base_directory = '/Volumes/Upwork/Upwork/Big_ticket/Data/Raw_data/'
```

```python
# Transverse through all directories and collect JSON files
def get_json_files(base_directory):
    json_files = {}
    # Traverse through all directories and collect JSON files
    for dirpath, dirnames, filenames in os.walk(base_directory):
        # Only proceed if 'uniqueIds_jsons' folder is in the directory path
        if 'uniqueIds_jsons' in dirpath:
            # Collect all .json files in this directory
            json_files[dirpath] = [os.path.join(dirpath, f) for f in filenames if f.endswith('.json')]
            
            # Debugging: Print JSON files collected and directory
            print(f"Found 'uniqueIds_jsons' folder: {dirpath}")
            print(f"Collected JSON Files: {json_files[dirpath]}")

            # If there are no JSON files in the folder, print a message
            if len(json_files[dirpath]) == 0:
                print(f"No JSON files found in {dirpath}")
            
    return json_files
```

```python
# Upload JSON files to MongoDB
def upload_jsons_to_mongo(json_files):
    folder_file_count = {}
    
    for folder, files in json_files.items():
        count = 0  # Counter for the number of files in each folder
        for json_file in files:
            try:
                # Open each JSON file and load its contents
                with open(json_file, 'r') as f:
                    data = json.load(f)
                    
                    # Insert data into MongoDB collection
                    collection.insert_one(data)
                    
                    count += 1
                    print(f"Uploaded: {json_file}")
            except Exception as e:
                print(f"Failed to upload {json_file}: {str(e)}")
        
        folder_file_count[folder] = count  # Track the number of files uploaded from this folder
    
    return folder_file_count
```

```python
# Get all JSON files from subfolders
json_files = get_json_files(base_directory)

# Debugging: Check the total number of files found
print(f"Total Folders Found: {len(json_files)}")
print(f"Total JSON Files Found: {sum(len(files) for files in json_files.values())}")

# Upload the JSON files to MongoDB and get the count of files uploaded from each folder
folder_file_count = upload_jsons_to_mongo(json_files)

# Print how many files were uploaded from each folder
for folder, count in folder_file_count.items():
    print(f"Folder: {folder} - Files Uploaded: {count}")
```
