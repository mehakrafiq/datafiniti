```python
import os
import json
import pandas as pd
import re
from collections import Counter
```

```python
# Define the base working directory
directory = '/Volumes/Upwork/Upwork/Big_ticket/Data/Raw_data/'

# Define the base working directory
base_directory = os.path.join(directory, 'Water_filtration_systems')
```

```python
# Define the output folder path where the JSON files are currently stored
output_folder_path = os.path.join(base_directory, 'renamed_jsons')
```

```python
# Create a dataframe to store all the json files

# List to store data for the DataFrame
data = []

# Iterate through all files in the output folder
for file_name in os.listdir(output_folder_path):
    if file_name.endswith('.json'):  # Process only JSON files
        file_path = os.path.join(output_folder_path, file_name)
        
        # Read and process the JSON file
        with open(file_path, 'r') as file:
            try:
                json_object = json.load(file)
                
                # Add the file name to the JSON object for reference
                json_object['file_name'] = file_name
                
                # Add the JSON object to the list
                data.append(json_object)

            except json.JSONDecodeError:
                # Handle or log the error if necessary
                pass

# Create a pandas DataFrame from the data list
df = pd.json_normalize(data)

# Print shape of the DataFrame
print(df.shape)

# Display the DataFrame
df.head()
```

```python
# make new df of ean and features
df2 = df[['id', 'features']]
df2.head()
```

```python
# Function to remove "value" key from each dictionary in the list
def remove_value_key(feature_list):
    if isinstance(feature_list, list):  # Only process if it's a list
        for feature_dict in feature_list:
            if isinstance(feature_dict, dict) and "value" in feature_dict:  # Only process if it's a dictionary
                del feature_dict["value"]
    # Return the original feature_list if not a list or no modification needed
    return feature_list

# Apply the function to the 'feature' column
df2['features'] = df2['features'].apply(remove_value_key)

# Display the modified DataFrame
df2.head()
```

```python
# Function to extract the value from 'key' in each dictionary
def extract_keys(feature_list):
    if isinstance(feature_list, list):  # Check if the input is a list
        # Extract 'key' from each dictionary, if the 'key' exists
        return [feature_dict["key"] for feature_dict in feature_list if isinstance(feature_dict, dict) and "key" in feature_dict]
    else:
        return None  # Ignore non-list entries by returning None

# Apply the function to the 'features' column
df2['features'] = df2['features'].apply(extract_keys)

# Display the modified DataFrame
df2.head()
```

```python
df2.shape
```

```python
# Step 1: Normalize the 'features' column by splitting the comma-separated values
df2['normalized_features'] = df2['features'].apply(
    lambda x: [feature.strip() for feature in x] if isinstance(x, list) else None
)

# Step 2: Explode the 'normalized_features' column to have one feature per row
df_exploded = df2.explode('normalized_features')

# Step 3: Create a DataFrame where each feature is a column and the values are the counts
# Use get_dummies to create a binary indicator DataFrame
df_dummies = pd.get_dummies(df_exploded['normalized_features'])

# Group by the original index (if needed, this step is optional based on how you want to structure the data)
df_feature_counts = df_dummies.groupby(df_exploded.index).sum()

# Display the resulting DataFrame
df_feature_counts.head()
```

```python
# Display the top 100 features
df_feature_counts.sum().sort_values(ascending=False).head(50)
```

```python
# List features that are present in at least 25% of the products
common_features = df_feature_counts.columns[df_feature_counts.mean() > 0.10].tolist()

#list of common features
print(common_features)
```

```python
print("Number of features:", len(common_features))
```

```python
# save list of top_features in a csv file

# Define the folder paths relative to the base directory
input_folder_path = os.path.join(base_directory, 'renamed_jsons')
output_folder_path = os.path.join(base_directory, 'top_features')

# Create the output directory if it doesn't exist
os.makedirs(output_folder_path, exist_ok=True)

# Make a dataframe of top features
df_top_features = pd.DataFrame(common_features, columns=['feature'])

# Save the top features to a CSV file
df_top_features.to_csv(os.path.join(output_folder_path, 'top_features.csv'), index=False)

```
