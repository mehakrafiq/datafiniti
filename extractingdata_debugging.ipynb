```python
# Import Libraries
import json
import os
import shutil
import pandas as pd
```

```python
# Define the base working directory
base_directory = '/Volumes/Upwork/Upwork/Big_ticket/test-dishwasher'

# Define the folder paths relative to the base directory
input_folder_path = os.path.join(base_directory, 'raw')
output_folder_path = os.path.join(base_directory, 'output_jsons')

# Create the output directory if it doesn't exist
os.makedirs(output_folder_path, exist_ok=True)

# Initialize a global counter for JSON objects
global_json_counter = 1

# Iterate through all files in the input folder
for file_name in os.listdir(input_folder_path):
    if file_name.endswith('.txt'):  # Process only text files
        file_path = os.path.join(input_folder_path, file_name)
        
        # Initialize a list to store JSON objects
        json_objects = []

        # Read and process the file
        with open(file_path, 'r') as file:
            for line in file:
                # Attempt to parse each line as JSON
                try:
                    json_object = json.loads(line)
                    json_objects.append(json_object)
                except json.JSONDecodeError:
                    # Handle or log the error if necessary
                    pass

        # Save each JSON object to a separate file with a global counter
        for json_object in json_objects:
            output_file_path = os.path.join(output_folder_path, f'json_object_{global_json_counter}.json')
            with open(output_file_path, 'w') as outfile:
                json.dump(json_object, outfile)
            
            # Increment the global counter
            global_json_counter += 1

        # Output the list of generated file paths
        generated_files = [os.path.join(output_folder_path, f'json_object_{i}.json') for i in range(global_json_counter - len(json_objects), global_json_counter)]
        print(f'Generated files for {file_name}:', generated_files)
```

```python
# Define the base working directory
base_directory = '/Users/mehak/Upwork/Big_ticket/test-dishwasher'

# Define the folder paths relative to the base directory
output_folder_path = os.path.join(base_directory, 'output_jsons')

# File paths for the two JSON objects
file_path_1 = os.path.join(output_folder_path, 'json_object_1.json')
file_path_2 = os.path.join(output_folder_path, 'json_object_1921.json')

# Load the JSON object from the first file
with open(file_path_1, 'r') as file:
    json_data_1 = json.load(file)

# Load the JSON object from the second file
with open(file_path_2, 'r') as file:
    json_data_2 = json.load(file)

# Function to recursively extract all keys from a JSON object
def extract_keys(obj, keys=set()):
    if isinstance(obj, dict):
        for key, value in obj.items():
            keys.add(key)
            extract_keys(value, keys)
    elif isinstance(obj, list):
        for item in obj:
            extract_keys(item, keys)
    return keys

# Extract all keys from both JSON data
all_keys_1 = extract_keys(json_data_1)
all_keys_2 = extract_keys(json_data_2)

# Find common keys
common_keys = all_keys_1.intersection(all_keys_2)

# Find unique keys in each JSON data
unique_keys_code1 = all_keys_1 - all_keys_2
unique_keys_code2 = all_keys_2 - all_keys_1

# Convert sets to lists
common_keys = list(common_keys)
unique_keys_code1 = list(unique_keys_code1)
unique_keys_code2 = list(unique_keys_code2)

# Find the maximum length among the three lists
max_length = max(len(common_keys), len(unique_keys_code1), len(unique_keys_code2))

# Pad the shorter lists with None to match the maximum length
common_keys.extend([None] * (max_length - len(common_keys)))
unique_keys_code1.extend([None] * (max_length - len(unique_keys_code1)))
unique_keys_code2.extend([None] * (max_length - len(unique_keys_code2)))

# Create a DataFrame to display the results
df = pd.DataFrame({
    'common_keys': common_keys,
    'unique_keys_code1': unique_keys_code1,
    'unique_keys_code2': unique_keys_code2
})
print(df)
```

```python
# Define the base working directory
base_directory = '/Users/mehak/Upwork/Big_ticket/test-dishwasher'

# Define the folder path for in_stock
in_stock_folder_path = os.path.join(base_directory, 'in_stock')

# Initialize counters for files with and without the 'prices' key
files_with_prices_key = 0
files_without_prices_key = 0
total_files = 0

# Iterate through all JSON files in the in_stock folder
for file_name in os.listdir(in_stock_folder_path):
    file_path = os.path.join(in_stock_folder_path, file_name)

    # Ensure we are only processing .json files and not directories or other file types
    if file_name.endswith('.json') and os.path.isfile(file_path):
        try:
            # Load the JSON object from the file
            with open(file_path, 'r') as file:
                json_data = json.load(file)
            
            total_files += 1

            # Check if the key 'prices' exists
            if 'prices' in json_data:
                files_with_prices_key += 1
            else:
                files_without_prices_key += 1

        except Exception as e:
            print(f"Error processing file {file_name}: {e}")

# Output the results
print(f"Total JSON files processed in 'in_stock' folder: {total_files}")
print(f"Number of files with 'prices' key: {files_with_prices_key}")
print(f"Number of files without 'prices' key: {files_without_prices_key}")
```
